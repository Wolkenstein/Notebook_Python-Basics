## Import of all needed Python libraries

```python
import pandas as pd

from nltk.corpus import stopwords
from nltk import bigrams
from wordcloud import WordCloud
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = "retina"
matplotlib.rcParams["figure.figsize"] = (13,8)
```

## First step: Loading the data

Read the .csv-table from [GitHub](https://github.com/hepplerj/whatisdigitalhumanities) into a Pandas Dataframe and display 5 randomly selected rows for testing.

```python
url = "https://raw.githubusercontent.com/hepplerj/whatisdigitalhumanities/master/dayofquotes_full.csv"
df = pd.read_csv(url)
display(df.sample(n=5))
texts = df.loc[:, "quote"].tolist()
```

## Load stopwords and add stopwords to list

```python
stopwords = set(stopwords.words('english'))
stopwords.add('DH')
stopwords.add('digital')
stopwords.add('humanities')
```

## Generate and render wordcloud

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white",
                      width=1280, height=720).generate(" ".join(texts))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
